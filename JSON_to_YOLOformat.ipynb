# Here the convert function helps to convert COCO format into YOLO text file format. 

# Input of the function are:
 
1.   An array which contain 2 things first is the class (Person or Car) and second is the boundary box details which contain top-left corner point, width and height of the boundary box.
2.   width of the Image
3.   Height of the Image

#Output of the functions is:


1.   An array which contain first values as a class (0 - Person, 1 - Car), center coordinate, width and height of the boundary box.

#All the values in the output array are normallized.


```python
def convert(arr,img_width,img_height):
  ans = []
  ans.append(arr[0]-1)
  bbox = arr[1]

  xmin = bbox[0]
  ymin = bbox[1]
  xmax = xmin + bbox[2]
  ymax = ymin + bbox[3]

  x_center = (xmin + xmax)/2
  y_center = (ymin + ymax)/2

  absolute_x = x_center/img_width
  absolute_y = y_center/img_height
  absolute_width = bbox[2]/img_width
  absolute_height = bbox[3]/img_height

  ans.append(absolute_x)
  ans.append(absolute_y)
  ans.append(absolute_width)
  ans.append(absolute_height)
  return ans
```

```python
b = [1,[846, 145, 146, 477]]
w,h =  1024,768

ans = convert(b,w,h)
print(ans)
```

```python
def write_txt_file(L,file_name):

  line = " ".join([str(a) for a in L]) + '\n'

  file1 = open(file_name+".txt","a")
  file1.writelines(line)
```

# This is the final code after running all the above codes. This code is giving the actuall output where it is storing all the annotations of the specific image into a text file of same name
